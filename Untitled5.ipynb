```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import csv


#dataset
from sklearn.datasets import load_digits
digitos= load_digits()
# separeação do dados
from sklearn.model_selection import train_test_split
x_treino,x_test,y_treino,y_test= train_test_split(digitos.data, digitos.target, test_size=1/7, random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(x_treino)

x_treino = scaler.transform(x_treino)
x_test = scaler.transform(x_test)
```

```python
from sklearn.decomposition import PCA

# Instanciar o modelo
pca_99 = PCA(0.99)

# Passar os dados de treino para o PCA
pca_99.fit(x_treino)

# Mapear os dados de treino e teste
x_treino_99 = pca_99.transform(x_treino)
print('Dimensao dos dados com 99 por cento de variabilidade', x_treino_99.shape)

x_test_99 = pca_99.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
import seaborn as sns
from sklearn.metrics import confusion_matrix, classification_report

RegressaoLogistica = LogisticRegression(solver='lbfgs')

# Treinar o modelo nos dados originais
RegressaoLogistica.fit(x_treino, y_treino)

# Prever os dados de teste
predic = RegressaoLogistica.predict(x_test)

cm = confusion_matrix(y_test, predic)
cr = classification_report(y_test, predic)
print(cr)
print(cm)
```

```python
# treinar modelos com dados reduzidos
# Treinar o modelo com dados reduzidos a 99%
RegressaoLogistica.fit(x_treino_99, y_treino)

# Prever os dados de teste
predic = RegressaoLogistica.predict(x_test_99)

cm_99 = confusion_matrix(y_test, predic)
cr_99 = classification_report(y_test, predic)
print(cr_99)
print(cm_99)
```
